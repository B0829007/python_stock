```python
import requests
import numpy as np
import pandas as pd
import json

ticket = "AAPL"
APIurl = "https://query1.finance.yahoo.com/v8/finance/chart/"+ticket+"?period1=1262304000&period2=1577836800&interval=1mo&events=history&=hP2rOschxO0"
response = requests.get(APIurl)  #取出資料json檔
data = json.loads(response.text)  #讀取文字檔呈現
table = pd.DataFrame(data['chart']['result'][0]['indicators']['quote'][0], index=pd.to_datetime(np.array(data['chart']['result'][0]['timestamp'], dtype='int64')*1000*1000*1000))
table
```

```python
def Strategy1():
    stock_sum=0
    balance=0
    for i in table.close:
        this_month_buying=1000//i+balance//i
        balance=1000%(i//1)+balance%(i//1)
        stock_sum+=this_month_buying
    total=table.close[-1]*stock_sum+balance
    print("Strategy1 total: ")
    print(total)

def Strategy2():
    stock_sum=0
    balance=0
    this_month_buying=1000//table.close[0]
    balance=1000%(table.close[0]//1)+balance%(table.close[0]//1)
    stock_sum+=this_month_buying
    for i in range(1,116):
        if(table.close[i]<table.close[i-1]):
            this_month_buying=1000//table.close[i]+balance//table.close[i]
            balance=1000%(table.close[i]//1)+balance%(table.close[i]//1)
            stock_sum+=this_month_buying
        else:
            this_month_buying=800//table.close[i]
            balance=800%(table.close[i]//1)+200+balance
            stock_sum+=this_month_buying
            
    for i in range (116,121):
        this_month_buying=1000//i+balance//i
        balance=1000%(i//1)+balance%(i//1)
        stock_sum+=this_month_buying
    
    total=table.close[-1]*stock_sum+balance
    print("Strategy2 total: ")
    print(total)
```

```python
Strategy1()
Strategy2()
```
